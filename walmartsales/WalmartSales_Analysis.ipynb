# 1. Import necess lib and Create Spark Session #

```
# import pyspark
import pyspark

# SparkSession
from pyspark.sql import SparkSession

# Import function
import pyspark.sql.functions as F
```

```
# Create SparkSession
spark = SparkSession.builder.master("local[*]") \
                            .appName("WalmartSales_Analysis") \
                            .getOrCreate()

# Print spark version
print("Spark version: ", spark.version)
```

```
# Read dataset into Spark Dataframe
    # Define file_path
file_path = "dbfs:/FileStore/tables/WalmartSalesData.csv"
    # Define file_type
file_type = "csv"

# Read data
walmartsales_df = spark.read.format(file_type) \
                            .option("header", True) \
                            .option("inferSchema", True) \
                            .load(file_path)

# Display dataframe
display(walmartsales_df)
```

```
# Create day, month, year column
walmartsales_df = walmartsales_df.withColumn("Day", F.day(F.col("Date"))) \
                                 .withColumn("Month", F.month(F.col("Date"))) \
                                 .withColumn("Year", F.year(F.col("Date")))

# Display df
display(walmartsales_df)
```

# 2. General analysis #

```
# 2.1. How many unique city in dataset?
unique_city = walmartsales_df.select("City").distinct()

# 2.2. How many NumberOfInvoice for each city in each branch?
numberinvoice_branch_city = walmartsales_df.groupBy(["City", "Branch"]).count()
```

```
# Show result of unique city
display(unique_city)
```

```
# Show result of NumberOfInvoice
display(numberinvoice_branch_city.orderBy("count", ascending = True))
```

# 3. Product Analysis

```
# 3.1. How many unique product lines does the data have?
unique_product = walmartsales_df.select("Product line").distinct()

# 3.2. What is the most common payment method?
most_common_payment_type = walmartsales_df.groupBy("payment").count() \
                                                             .withColumnRenamed("count", "NumberOfUsingPaymentType") \
                                                             .orderBy("NumberOfUsingPaymentType", ascending = False)
# 3.3. What is the most selling product line?
most_sell_product = walmartsales_df.groupBy("Product line").agg({"Quantity" : "count"}) \
                                                           .withColumnRenamed("count(Quantity)", "NumberOfSellProduct") \
                                                           .orderBy("NumberOfSellProduct", ascending = False)

# 3.4. What is the total revenue by month?
total_revenue = walmartsales_df.groupBy("Month").agg({"Total": "sum"}) \
                                                .withColumnRenamed("sum(Total)", "TotalRevenuebyMonth") \
                                                .orderBy("Month", ascending = True)
# 3.5. What month had the largest COGS?
largest_cogs = walmartsales_df.groupBy("Month").agg({"cogs": "sum"}) \
                                               .withColumnRenamed("sum(cogs)", "Largest_Cogs") \
                                               .orderBy("Largest_Cogs", ascending = False)
```

```
# 3.6. What product line had the largest revenue? 
product_largest_revenue = walmartsales_df.groupBy("Product line").agg({"Total": "sum"}) \
                                                                 .withColumnRenamed("sum(Total)", "TotalRevenue") \
                                                                 .orderBy("TotalRevenue", ascending = False)

# 3.7. What is the city with the largest revenue?
city_largest_revenue = walmartsales_df.groupBy("City").agg({"Total": "sum"}) \
                                                      .withColumnRenamed("sum(Total)", "TotalRevenue") \
                                                      .orderBy("TotalRevenue", ascending = False)

# 3.8. What product line had the largest VAT?
product_largest_vat = walmartsales_df.groupBy("Product line").agg({"Tax 5%": "avg"}) \
                                                             .withColumnRenamed("avg(Tax 5%)", "Largest VAT") \
                                                             .orderBy("Largest VAT", ascending = False)

# 3.9. Fetch each product line and add a column to those product line showing "Good", "Bad". Good if its greater than average sales
average_sales = walmartsales_df.agg({"Quantity": "avg"}).collect()[0][0]
fetch_product_status = walmartsales_df.groupBy("Product line").agg({"Quantity": "avg"}) \
                                                         .withColumnRenamed("avg(Quantity)", "AVG_Sales_Product") \
                                                         .withColumn("Status_Product", F.when(F.col("AVG_Sales_Product") > average_sales, "Good").otherwise("Bad")) \
                                                         .select(["Product line", "Status_Product"])

# 3.10. Which branch sold more products than average product sold?
branch_sold_avg = walmartsales_df.groupBy("Branch").agg({"Quantity": "sum"}) \
                                                   .withColumnRenamed("sum(Quantity)", "SUM_Branch_Sales") \
                                                   .filter(F.col("SUM_Branch_Sales") > average_sales) \
                                                   .orderBy("SUM_Branch_Sales", ascending = True)

# 3.11. What is the most common product line by gender?
most_product_gender = walmartsales_df.groupBy(["Gender", "Product line"]).agg({"Product line": "count"}) \
                                                       .withColumnRenamed("count(Product line)", "NumberofProductSold") \
                                                       .orderBy("NumberofProductSold", ascending = False)
# 3.12. What is the average rating of each product line?
average_rating_product = walmartsales_df.groupBy("Product line").agg({"Rating": "avg"}) \
                                                                .withColumnRenamed("avg(Rating)", "AVG_Rating") \
                                                                .orderBy("AVG_Rating", ascending = False)
```
